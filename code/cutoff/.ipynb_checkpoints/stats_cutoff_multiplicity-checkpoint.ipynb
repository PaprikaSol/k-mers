```r
# Load libraries
library(stringr)
library(glue)
library(purrr)
# library(plotly)
# library(tictoc)
```

```r
#Get the statistical results for a specific dataset and cutoff - [1]
stats_cutoff_multiplicity<- function(data){
    
    wilcoxon <- matrix(, nrow = 3, ncol = 9, dimnames = list(c('Low-High', 'Low-Total', 'High-Total'), seq(100, 900, 100)))
    fisher <- matrix(, nrow = 3, ncol = 9, dimnames = list(c('Low-High', 'Low-Total', 'High-Total'), seq(100, 900, 100)))
    kolm_smirnov <- matrix(, nrow = 3, ncol = 9, dimnames = list(c('Low-High', 'Low-Total', 'High-Total'), seq(100, 900, 100)))
    
    #STATISTICAL TESTS [WILCOXON, FISHER, KOLMOGOROV-SMIRNOV]  

    for (cut in 1:9){
        
        low_mult <- data[data$Value < 100*cut, ]
        high_mult <- data[data$Value >= 100*cut, ]

        low = table(low_mult[, 2])
        high = table(high_mult[, 2])
        total = table(data[, 2])
    
        #Wilcoxon Test
        wilcoxon[1, cut] = wilcox.test(low, high)$p.value
        wilcoxon[2, cut] = wilcox.test(low, total)$p.value
        wilcoxon[3, cut] = wilcox.test(high, total)$p.value

        #Fisher Test
        fisher[1, cut] = var.test(low, high)$p.value
        fisher[2, cut] = var.test(low, total)$p.value
        fisher[3, cut] = var.test(high, total)$p.value

        #Kolmogorov-Smirnov
        kolm_smirnov[1, cut] = ks.test(low, high)$p.value
        kolm_smirnov[2, cut] = ks.test(low, total)$p.value
        kolm_smirnov[3, cut] = ks.test(high, total)$p.value
        
#         #Shapiro
#         saphiro[1, cut] = .test(low, high)$p.value
#         saphiro[2, cut] = cor.test(low, total)$p.value
#         shapiro[3, cut] = cor.test(high, total)$p.value
        }

    return (list(wilcoxon, fisher, kolm_smirnov))
}
```

```r
#Main - [1]
options(warn=-1)

#Choose k
k = 20

#Load data
data_path = file.path('C:','Users', 'user', 'Desktop', 'Workspaces', 'R', 'INAB', 'data', glue('sars_1000_info_k={k}.txt'))
data <- read.table(data_path)

#Filter data for 'ACTG'
filt_data <- data[str_detect(data[, 1], '^[ACTG]+$'),]

stats = stats_cutoff_multiplicity(filt_data)
stats

#Save results
res_path = file.path('C:','Users', 'user', 'Desktop', 'Workspaces', 'R', 'INAB', 'results')
write.csv(stats[1], glue('{res_path}/stats_multiplicity_wilcox_k={k}'))
write.csv(stats[2], glue('{res_path}/stats_multiplicity__fisher_k={k}'))
write.csv(stats[3], glue('{res_path}/stats_multiplicity__kolm_k={k}'))
```

```r
#Plots - ALL [1]
img_path = glue('{res_path}/img')
# png(glue('{img_path}/stats_multiplicity_k={k}.png'))
par(mfcol=c(3,3))


x_axis = 100*(1:9)
y_lim = c(0, 1)
# y_lim = NULL

#Wilcoxon
df_w = as.data.frame(stats[1])

plot(x =  x_axis, y = unlist(df_w[1, ]), type = 'o', main = 'Wilcoxon : Low-High', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

plot(x =  x_axis, y = unlist(df_w[2, ]), type = 'o', main = 'Wilcoxon : Low-Total', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

plot(x =  x_axis, y = unlist(df_w[3, ]), type = 'o', main = 'Wilcoxon : High-Total', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

#Fisher
df_f = as.data.frame(stats[2])

plot(x = x_axis, y = unlist(df_f[1, ]), type = 'o', main = 'Fisher : Low-High', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

plot(x = x_axis, y = unlist(df_f[2, ]), type = 'o', main = 'Fisher : Low-Total', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

plot(x = x_axis, y = unlist(df_f[3, ]), type = 'o', main = 'Fisher : High-Total', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

#Kolmogorov-Smirnov
df_k = as.data.frame(stats[3])

plot(x = x_axis, y = unlist(df_k[1, ]), type = 'o', main = 'Kolm-Smirnov : Low-High', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

plot(x = x_axis, y = unlist(df_k[2, ]), type = 'o', main = 'Kolm-Smirnov : Low-Total', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

plot(x = x_axis, y = unlist(df_k[3, ]), type = 'o', main = 'Kolm-Smirnov : High-Total', xlab = 'cutoff', ylab = 'p-value', ylim = y_lim)
abline(h = 0.05)

# dev.off()
```

```r
#Get the statistical results for a specific dataset and cutoff [2]
stats_cutoff_mult_focused<- function(data, points){
    
    # Default = 10 points
    if(missing(points)){
        num_of_cuts = 9
    }
    
    min_mult = min(data$Value)
    max_mult = max(data$Value)
    cuts = points + 2 - 1
    seg = (max_mult - min_mult)/cuts
    seq_cuts = seg*(1:points)    
    
    wilcoxon <- matrix(, nrow = 3, ncol = points, dimnames = list(c('Low-High', 'Low-Total', 'High-Total'), seq_cuts))
    fisher <- matrix(, nrow = 3, ncol = points, dimnames = list(c('Low-High', 'Low-Total', 'High-Total'), seq_cuts))
    kolm_smirnov <- matrix(, nrow = 3, ncol = points, dimnames = list(c('Low-High', 'Low-Total', 'High-Total'), seq_cuts))
    
    #STATISTICAL TESTS [WILCOXON, FISHER, KOLMOGOROV-SMIRNOV]  
    
    for (cut in 1:points){
        
        low_mult <- data[data$Value < seq_cuts[cut], ]
        high_mult <- data[data$Value >= seq_cuts[cut], ]

        low = table(low_mult[, 2])
        high = table(high_mult[, 2])
        total = table(data[, 2])
    
        #Wilcoxon Test
        wilcoxon[1, cut] = wilcox.test(low, high)$p.value
        wilcoxon[2, cut] = wilcox.test(low, total)$p.value
        wilcoxon[3, cut] = wilcox.test(high, total)$p.value

        #Fisher Test
        fisher[1, cut] = var.test(low, high)$p.value
        fisher[2, cut] = var.test(low, total)$p.value
        fisher[3, cut] = var.test(high, total)$p.value

        #Kolmogorov-Smirnov
        kolm_smirnov[1, cut] = ks.test(low, high)$p.value
        kolm_smirnov[2, cut] = ks.test(low, total)$p.value
        kolm_smirnov[3, cut] = ks.test(high, total)$p.value
        }

    return (list(wilcoxon, fisher, kolm_smirnov))
}
```

```r
#Main [2]
options(warn=-1)

#Choose k
k = 20
points = 24

#Load data
data_path = file.path('C:','Users', 'user', 'Desktop', 'Workspaces', 'R', 'INAB', 'data', glue('sars_1000_info_k={k}.txt'))
data <- read.table(data_path)

#Filter data for 'ACTG'
filt_data <- data[str_detect(data[, 1], '^[ACTG]+$'),]

stats = stats_cutoff_mult_focused(filt_data, points)
stats

#Save results
res_path = file.path('C:','Users', 'user', 'Desktop', 'Workspaces', 'R', 'INAB', 'results')
write.csv(stats[1], glue('{res_path}/stats_multiplicity_wilcox_k={k}'))
write.csv(stats[2], glue('{res_path}/stats_multiplicity__fisher_k={k}'))
write.csv(stats[3], glue('{res_path}/stats_multiplicity__kolm_k={k}'))
```
